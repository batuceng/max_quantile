```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture

# Step 1: Generate Data
np.random.seed(42)

# Generate samples from N(1, 3)
data1 = np.random.normal(loc=0.75, scale=0.05, size=11000)

# Generate samples from N(4, 1)
data2 = np.random.normal(loc=0, scale=0.002, size=1)

# Generate samples from N(7, 2)
data3 = np.random.normal(loc=-0.75, scale=0.05, size=9000)

# Combine the data
data = np.concatenate([data1, data2, data3])

# Step 2: Fit a Gaussian Mixture Model
gmm = GaussianMixture(n_components=3, random_state=42)
gmm.fit(data.reshape(-1, 1))

# Step 3: Visualize the Results
# Generate points for plotting the GMM
x = np.linspace(-1, 1, 1000).reshape(-1, 1)
logprob = gmm.score_samples(x)
pdf = np.exp(logprob)
responsibilities = gmm.predict_proba(x)
pdf_individual = responsibilities * pdf[:, np.newaxis]


# Plot the data histogram
bin_count = 50
# hist = plt.hist(data, bins=bin_count, density=True, alpha=0.6, label='Data')
hist = np.histogram(data, bins=bin_count, density=True)
protos = (hist[1][1:]+hist[1][:-1])/2
digital_labels = np.fmin(np.digitize(data, hist[1]), bin_count)

densities = hist[0]
areas = hist[1][1:] - hist[1][:-1]
probs = densities * areas


sorted_indices = densities.argsort()[::-1]
sorted_probs = probs[sorted_indices]
sorted_cumsum = sorted_probs.cumsum()
index = (sorted_cumsum < 0.50).cumsum().argmax()
density_th = densities[index+1]

plt.figure()

left = hist[1][:-1]
right = hist[1][1:]
left_min = np.min(left)
right_max = np.max(right)
legend_red, legend_blue = True, True
for i in range(len(protos)):
    index_density = densities[i]
    index_prob = probs[i]
    index_left = left[i]
    index_right = right[i]
    if index_density > density_th:
        plt.bar((index_left+index_right)/2, index_density, width=index_right-index_left, color='red', alpha=0.6, label='covered' if legend_red else None)
        legend_red = False
    else:
        plt.bar((index_left+index_right)/2, index_density, width=index_right-index_left, color='blue', alpha=0.6, label='uncovered' if legend_blue else None)
        legend_blue = False

# Plot the GMM components
plt.scatter(protos, [0]*bin_count, marker='o', color='red', label='prototypes', zorder=5)
plt.plot(x, pdf, '-k', label='True Dist')
plt.hlines(y=density_th+0.08, xmin=-1, xmax=1, linestyles='--', colors='black')
# plt.plot(x, pdf_individual, '--k')

# Plot individual Gaussian components
for i in range(3):
    # plt.plot(x, pdf_individual[:, i], label=f'Component {i+1}')
    pass

plt.xlabel('Value y')
plt.ylabel('Density p(y|x)')
plt.title(f'p(y|x) Bimodal, Coverage: {100*sorted_cumsum[index+1]:.2f}%, #bins: {bin_count}')
plt.legend()
plt.show()
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F

class UnconditionalClassifier(nn.Module):
    def __init__(self, in_size = 1, no_of_classes=50):
        super(UnconditionalClassifier, self).__init__()
        self.in_size = in_size
        self.no_of_classes = no_of_classes
        self.hidden_size = 16
        self.net = nn.Sequential(
            nn.Linear(self.in_size, self.hidden_size),
            nn.ReLU(),
            nn.Linear(self.hidden_size, self.hidden_size),
            nn.ReLU(),
            nn.Linear(self.hidden_size, self.no_of_classes),
        )
        
    def forward(self, x):
        x = self.net(x)
        return x
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F

class UnconditionalProtoClassifier(nn.Module):
    def __init__(self, in_size = 1, no_of_classes=50):
        super(UnconditionalProtoClassifier, self).__init__()
        self.in_size = in_size
        self.no_of_classes = no_of_classes
        self.hidden_size = 16
        self.net = nn.Sequential(
            nn.Linear(self.in_size, self.hidden_size),
            nn.ReLU(),
            nn.Linear(self.hidden_size, self.hidden_size),
            nn.ReLU(),
            nn.Linear(self.hidden_size, self.no_of_classes),
        )
        
    def forward(self, x):
        x = self.net(x)
        return x
```

```python
# Simple demo with prototype centers and hard/soft labels
y_true = 0.2
probo_boundaries = np.linspace(0,1,51)
# probo_boundaries = np.array([0., 0.2, 0.21, 0.22, 0.3,0.5, 0.55, 1.0])
probo_centers = (probo_boundaries[1:]+probo_boundaries[:-1])/2
widths = (probo_boundaries[1:] - probo_boundaries[:-1])*0.75
probo_centers
```

```python
# Hard Labels
probo_values_hard = np.zeros_like(probo_centers)
probo_values_hard[np.argmin(np.abs(probo_centers-y_true))] = 1

plt.figure()
plt.title(f'Hard Labels, y={y_true}')
plt.bar(probo_centers, probo_values_hard, width=widths)
plt.scatter(probo_centers, np.zeros_like(probo_centers)+0.01, marker='x', c='r')
plt.show()
```

```python
# Soft Labels (Gumbel Softmax)
logits = -(np.abs(probo_centers-y_true)+1e-16)
logits = logits - logits.max()

probo_values_soft = np.exp(logits/0.05)
probo_values_soft = probo_values_soft/probo_values_soft.sum()

plt.figure()
plt.title(f'Soft Labels, y={y_true}')
plt.bar(probo_centers, probo_values_soft, width=widths)
plt.scatter(probo_centers, np.zeros_like(probo_centers)+0.01, marker='x', c='r')
plt.show()
```
